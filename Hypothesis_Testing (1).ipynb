##**QUESTION--> 1**

##What is a null hypothesis (H₀) and why is it important in hypothesis testing?


#**ANSWER -->**

* Definition: The null hypothesis ($H_{0}$) is a statement of "no effect," "no difference," or "status quo". It represents the claim that is assumed to be true unless there is strong evidence to the contrary.  

* Importance: It serves as the baseline for mathematical testing. We do not try to "prove" the null hypothesis; rather, we determine if the data provides enough evidence to reject it in favor of an alternative.

##**QUESTION--> 2**

##What does the significance level (α) represent in hypothesis testing?

#**ANSWER -->**

* The significance level ($\alpha$) is the probability of rejecting the null hypothesis when it is actually true.  

* It is the threshold for "strong evidence." Common values are $0.05$ (5%) or $0.01$ (1%).

----------------------------------------------------------------------
##**QUESTION--> 3**

##Differentiate between Type I and Type II errors.

#**ANSWER -->**

* Type I Error ($\alpha$): Rejecting the null hypothesis when it is actually true (a "false positive").  

* Type II Error ($\beta$): Failing to reject the null hypothesis when it is actually false (a "false negative").

##**QUESTION--> 4**

##Explain the difference between a one-tailed and two-tailed test. Give an example of each.

#**ANSWER -->**

* One-tailed Test: Used when the research question looks for a change in a specific direction (e.g., "Is the new drug better than the old one?").  


* Two-tailed Test: Used when looking for any difference, regardless of direction (e.g., "Is there any difference in weight between these two groups?").

##**QUESTION--> 5**
##A company claims that the average time to resolve a customer complaint is 10 minutes.A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At α = 0.05, test the claim.

#**ANSWER -->**

* Hypotheses: $H_{0}: \mu = 10$; $H_{a}: \mu \neq 10$.

* Data: $n = 9$, $\bar{x} = 12$, $s = 3$, $\alpha = 0.05$, $df = 8$.  

* Calculation: Using the t-test formula $t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$:$$t = \frac{12 - 10}{3 / \sqrt{9}} = \frac{2}{1} = 2.0$$

**Decision: The critical value for $t_{0.025, 8}$ is approximately $2.306$. Since $2.0 < 2.306$, we fail to reject the null hypothesis. There is not enough evidence to disprove the claim.**

##**QUESTION--> 6**

## When should you use a Z-test instead of a t-test?

#**ANSWER -->**

* Z-test: Use when the population standard deviation ($\sigma$) is known and the sample size is large ($n > 30$).  

* t-test: Use when the population standard deviation is unknown and the sample size is small ($n < 30$), relying instead on the sample standard deviation ($s$).
